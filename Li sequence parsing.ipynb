```python
#write a function to deal with gaps in the Li sequence. input is start/stop, read in sequence, count gap chars, add that to stop, get fresh sequence
def gap_parse(start, end, seq = Li_struct):
    #print("initial pre parsed length is ", len(seq[start:end]))
    #print("initial pre parsed seq is ", seq[start:end])
    seq_list = list(seq[start:end])
    gap_count = 0
    for character in seq_list:
        if character == "-":
            gap_count += 1
    print("gap count is ", gap_count)
    new_end = end + gap_count
    new_seq = seq[start:new_end]
    seq_out = new_seq.replace("-", "")
    #print("new sequence range is ", new_seq)
    #print("parsed length is ", len(seq_out))
    #print("parsed sequence is ", seq_out)
    return(seq_out)
    
    
```
